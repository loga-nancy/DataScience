```python
"""Use pandas to sort data frame ascending order with most calorie density"""
import pandas as pd
raw_data={
    'food':['barcon','strawberry','banna','spinatch'],
    'grams':[50,200,100,200],
    'calories':[271,64,89,46]
}

df=pd.DataFrame(raw_data, columns=['food','grams','calories'])
df
```

```python
df['calorie_dense']=df['calories']/df['grams']

df.sort_values('calorie_dense',ascending=False)
```
